```python
#1. How many people have French 'fr' as their Language? 
#a. First show all columns then all languages 
df['Language'].count
```

```python
#b. Check whether 'fr' is there or not 
df.loc[(df['Language']=='fr')].count()[0]
```

```python
#Compare Languages to fr
fr = df.loc[(df['Language'] == 'fr')].count()[0]
el = df.loc[(df['Language'] == 'el')].count()[0]
it = df.loc[(df['Language'] == 'it')].count()[0]
zh = df.loc[(df['Language'] == 'zh')].count()[0]
de = df.loc[(df['Language'] == 'de')].count()[0]
es = df.loc[(df['Language'] == 'es')].count()[0]
labels = ['fr', 'el','it','zh','de','es']
colors = ['blue', 'red','yellow','green','orange','black']
plt.pie([fr, el,it,zh,de,es], labels=labels, colors=colors, autopct='%.2f')

plt.title('French vs. Other Languages')
plt.show()
```
